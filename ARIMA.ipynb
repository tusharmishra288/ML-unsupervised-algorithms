```python
from statsmodels.tsa.seasonal import seasonal_decompose
d=seasonal_decompose(r_test_logscale)
trend=d.trend
seasonal=d.seasonal
residual=d.resid
```

```python
plt.subplot(411)
plt.plot(r_test_logscale)
plt.subplot(412)
plt.plot(trend,c='r')
plt.subplot(413)
plt.plot(seasonal,c='green')
plt.subplot(414)
plt.plot(residual,c='orange')
```

```python
from statsmodels.tsa.stattools import acf,pacf
r_acf=acf(r_test_logscale,nlags=20)
r_pacf=pacf(r_test_logscale,nlags=20,method='ols')
```

```python
plt.subplot(121)
plt.plot(r_acf)
plt.axhline(y=0,linestyle='--')
plt.title('acf')
plt.subplot(122)
plt.plot(r_pacf)
plt.title('pacf')
plt.axhline(y=0,linestyle='--')
```
